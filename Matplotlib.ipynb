```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Company": ["TCS", "Cisco", "Wipro", "Deloitte"],
    "Sales_2023": [100, 150, 120, 180],
    "Sales_2024": [130, 170, 140, 200]
}

df = pd.DataFrame(data)

# Plot
plt.plot(df["Company"], df["Sales_2023"])
plt.plot(df["Company"], df["Sales_2024"])

plt.xlabel("Company Name")
plt.ylabel("Sales (in Crores)")
plt.title("Company Sales Comparison (2023 vs 2024)", loc="left")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(5)
b = [2,4,6,8,10]
c = [5,6,7,8,9]

fig = plt.figure()
ax = plt.subplot()

ax.plot(a, b, label='Frequency')
ax.plot(a, c, label='Period')

ax.set_title("Frequency vs Period Graph")   # 👈 Title add
ax.legend(loc='upper left')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(5)
b = [2,4,6,8,10]
c = [5,6,7,8,9]
 
fig = plt.figure()
ax = plt.subplot()

ax.plot(a, b, 'k--', label='Frequency')   # dashed line
ax.plot(a, c, 'k-', label='Period')       # normal line

ax.set_title("Frequency vs Period Graph")
ax.legend(loc='upper left', fontsize='xx-large')

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
rainfall = [20,25,30,50,80,120,200,180,140,90,40,25]

plt.plot(months, rainfall, marker='o')

plt.xlabel("Months")
plt.ylabel("Rainfall (mm)")
plt.title("Months vs Rainfall Graph")

plt.show()
```

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]
y1 = [10,20,30,40,50]

x2 = [1,2,3,4,5]
y2 = [15,25,35,45,55]

plt.scatter(x1, y1, color='blue', label='Group 1')
plt.scatter(x2, y2, color='red', label='Group 2')

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot with Two Colors")

plt.legend()
plt.show()
```
