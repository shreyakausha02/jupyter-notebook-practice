```python
#CONCATENATION
import numpy as np
n1=np.array([5,10,15,20])
n2=np.array([25,30,35,40])
print("Iteration")
for a in n1:
    print(a)
print("Iteration")
for a in n2:
    print(a)
Concate=np.concatenate((n1,n2))
print(Concate)
    
```

```python
n1=np.array([5,10,15,20])
n2=np.array([25,30,35,40])
print("Iteration")
for a in n1:
    print(a)
print("Iteration")
for a in n2:
    print(a)
Concate1=np.stack((n1,n2))
Concate2=np.vstack((n1,n2))
Concate3=np.hstack((n1,n2))
Concate4=np.dstack((n1,n2))
Concate4=np.column_stack((n1,n2))
print(Concate1)
print(Concate2)
print(Concate3)
print(Concate4)
```

```python
#SEARCHING FOR A VALUE IN NUMPY
n1=np.array([5,10,15,20])
print("Iteration")
for a in n1:
    print(a)
    output=np.where(n1==15)
    print(output)
```

```python
#TO FIND MINIMUM ELEMENTS IN HORIZONAL AND VERTICAL AXIS
s=np.array([[23,50,60,90],[100,20,12,233],[234,111,389,567]])
print("ITERATING 2D ARRAY....")
print("Iteration")
for a in s:
    print(a)
    print("MINIMUM VALUE....\n: ",s.min())
    print("MINIMUM VALUE IN VERTICAL AXIS....\n: ",s.min(axis=0))
    print("MINIMUM VALUE IN HORIZONTAL AXIS....\n: ",s.min(axis=1))
```

```python
#INTERSECTION
n1=np.array([5,10,15,20,25])
n2=np.array([25,30,35,40,15])
print("Iteration")
for a in n1:
    print(a)
print("Iteration")
for a in n2:
    print(a)
    v=np.intersect1d(n1,n2)
    print(v)
```

```python
n1=np.array([5,10,15,20,30,35])
n2=np.array([25,30,35,40,5,10])
print("Iteration")
for a in n1:
    print(a)
print("Iteration\n")
for a in n2:
    print(a)
    t=np.setdiff1d(n1,n2)
    print("The Uncommon elements : ",t)
```

```python
#SET DIFFERENCE
n1=np.array([5,10,15,20])
n2=np.array([25,30,35,40])
t=np.setdiff1d(n1,n2)
print("The Uncommon elements : ",t)
```
