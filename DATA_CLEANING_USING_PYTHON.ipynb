```python
#CLEANING USING RANGE
marks=np.array([85,78,-5,120,98,45])
valid_marks=np.arange(0,101)
final_marks=np.where(np.isin(marks,valid_marks),marks,np.nan)
print(final_marks)
```

```python
#MASKING(advanced slicing)
age=np.array([32,-5,101,10,2,-20])
cleaned_age=np.where(age<0,'NOT VALID',age)
print(cleaned_age)
```

```python
data=np.array([10,20,30,40,50,np.nan,70])
indices=np.arange(len(data))
valid_indices=indices[~np.isnan(data)] #[f,f,f,f,f,t,f]->after applying (~)->[t,t,t,t,t,f,t]

# masking(works with boolean only)
cleaned_data=data[valid_indices] # data[t,t,t,t,t,f,t] ->whatever is true it gives in print

print(cleaned_data)
```
